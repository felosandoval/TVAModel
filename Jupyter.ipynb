## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla Universos

```python
%%sql

Create table universos(
universo VARCHAR(45) PRIMARY KEY,
admin_universo VARCHAR(45) NOT NULL,
evento_nexus TIMESTAMP
);
```

### Tabla Líneas Temporales

```python
%%sql

Create table lineas_temporales(
linea_temporal int PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
universo varchar(45) NOT NULL,
FOREIGN KEY (universo) references universos(universo)
);
```

### Tabla Dispositivos Temporales

```python
%%sql

Create table dispositivos_temporales(
id_dispositivo int PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
fecha_fabricacion TIMESTAMP NOT NULL
);
```

### Tabla Departamentos

```python
%%sql

Create table departamentos(
id_departamento int PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
nombre_departamento varchar(45) NOT NULL
);
```

### Tabla Historial Misiones

```python
%%sql

Create table historial_misiones(
id_mision int PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
id_departamento int,
fecha TIMESTAMP,
FOREIGN KEY (id_departamento) references departamentos(id_departamento)
);
```

### Tabla Empleados

```python
%%sql

create table empleados(
id_empleados int PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
nombre varchar(45) NOT NULL,
linea_temporal int NOT NULL,
id_departamento int NOT NULL,
id_dispositivo_temporal int NOT NULL,
cargo_departamento varchar(45),
FOREIGN KEY (linea_temporal) references lineas_temporales(linea_temporal),
FOREIGN KEY (id_departamento) references departamentos(id_departamento),
FOREIGN KEY (id_dispositivo_temporal) references dispositivos_temporales(id_dispositivo)
);
```

### Consulta 1
**Obtenga el nombre de todos los empleados del universo X.**

Ejemplo:

| Nombre       |
|--------------|
| Keanu Reeves |
| Rick Astley  |
| ...          |

**Solución:** _Selecciona el nombre de los empleados desde la tabla empleados y une con un join la tabla lineas temporales 
cuando la linea temporal del empleado sea igual a la de la tabla lineas temporales y donde el 
universo se llame de tal manera._

```python
%%sql

SELECT
    empleados.nombre
FROM
    empleados
JOIN
    lineas_temporales
ON
    empleados.linea_temporal = lineas_temporales.linea_temporal
WHERE
    lineas_temporales.universo = 'Earth-982';
```

### Consulta 2
**Obtenga la cantidad de líneas temporales por universo ordenadas de mayor a menor cantidad.**

Ejemplo:


| Universo     | Cantidad |
|--------------|----------|
| Earth-555326 |    9     |
| EarthDia     |    7     |
| ...          |   ...    |

**Solución:** _Selecciona el universo y cuantas veces existe desde las lineas temporales y une la tabla con un 
join universo cuando el universo de la linea temporal es igual al universo de universos 
despues agrupa las lineas temporales y las ordena descendentemente._

```python
%%sql
SELECT
    lineas_temporales.universo, count(lineas_temporales.linea_temporal) as cantidad
FROM 
    lineas_temporales
JOIN 
    universos
ON
    lineas_temporales.universo = universos.universo
    
GROUP BY lineas_temporales.universo

ORDER BY cantidad DESC
```

### Consulta 5
**Obtenga todas las misiones del año X de un departamento Y.**

._

```python
%%sql
SELECT
    historial_misiones.id_mision,departamentos.nombre_departamento as Departamento,date(historial_misiones.fecha) as fecha
FROM 
    historial_misiones
JOIN 
    departamentos
ON 
    historial_misiones.id_departamento = departamentos.id_departamento
WHERE
    departamentos.nombre_departamento = 'Forenses Temporales' AND
    historial_misiones.fecha BETWEEN '20210101' AND '20220101';
   
```

### Consulta 8
**Obtenga la id y nombre universo con más cantidad de empleados de un departamento X.**

Ejemplo:

|Id |Universo |
|-|-------|
| 12 |Earth-65 |

**Solución:** _Selecciona el id de departamento con las lineas temporales de universo une la tabla lineas temporales cuando
luego se igualan los empleados de linea temporal y linea temporal de lineas temporales. volvemos a unir otra tabla
en este caso seria empleados, entonces si el id de departamentos es igual al id de empleados, 
entonces filtrara por "X" departamento agrupandolo por departamento, ordenandolo descendentemente y trayendo el primero.
._

```python
%%sql
SELECT
    departamentos.id_departamento AS id,lineas_temporales.universo AS Universo
FROM 
    departamentos, empleados
JOIN
    lineas_temporales
ON
    empleados.linea_temporal=lineas_temporales.linea_temporal

WHERE 
    departamentos.id_departamento=empleados.id_departamento
    AND departamentos.nombre_departamento = 'Exploradores'


GROUP BY departamentos.id_departamento,lineas_temporales.universo

ORDER BY COUNT(id_empleados) DESC

LIMIT 1
```
